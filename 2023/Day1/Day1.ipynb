```python
import numpy as np
import re
```

```python
with open('C:\Files\Code\AdventOfCode23\Day1\Day1Input.txt') as file:
    data = [i for i in file.read().strip().split("\n")]
```

```python
dict = {
    "one":"1ne",
    "two":"2wo",
    "three":"3hree",
    "four":"4our",
    "five":"5ive",
    "six":"6ix",
    "seven":"7even",
    "eight":"8ight",
    "nine":"9ine",
}
```

```python
for word, number in dict.items():
    for x in range(0, len(data)):
        data[x] = data[x].replace(word, number)
```

```python
numbers=[]
for line in data:
    arrayx = re.findall('\d+', line)
    numbers.append(arrayx)
```

```python
total=0
run=[]
for y in numbers:
    y = ''.join(y)
    print(len(str(y)))
    if len(str(y)) == 1:
        num = int(y[0] + y[0])
    else:
        num = int(str(y[0]) + str(y[-1]))
    total += num
    run.append(num)
```
